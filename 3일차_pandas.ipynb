### 통계함수
* 열이름.통계함수(): 열 이름이 여러 개면 리스트화

```python
# 중간고사, 기말고사 평균 계산

meandata=score1[['중간고사','기말고사']].mean()
meandata
```

```python
type(meandata) # 즉 중간고사 기말고사는 인덱스!
```

### 그룹화
- 변수명.groupby('그룹열이름')['열이름'].통계함수()
  - 그룹화한 열의 통계 분포를 알려 줌
  - 그룹 열 이름: 그룹화할 열의 이름
  - 열 이름: 그룹별로 통계 데이터를 확인하고자 하는 열의 이름
- 변수명.groupby(['그룹열이름1','그룹열이름2'])[['열이름1','열이름2']].통계함수()

```python
# 분반을 기준으로 기말고사 평균 계산

score1.groupby('분반')['기말고사'].mean()
```

```python
# 분반, 성별 기준으로 중간고사, 기말고사 평균 계산

score1.groupby(['분반','성별'])[['중간고사','기말고사']].mean()
```

### 데이터 결합
* 열 방향: 변수명=pd.merge(변수명1,변수명2,on='결합기준열이름',how='결합방향')
  - 2개의 테이블을 열 방향으로 결합
  - 변수명1, 변수명2: DataFrame 구조를 갖는 테이블
  - 결합하고자 하는 테이블들은반드시 동일한 열 이름이 존재해야 함
  - on='결합기준열이름': 두 개의 테이블을 결합할 때 기준이 되는 열 이름
    - 결합 기준 열의 데이터가 일치하는 부분이 없다면 결합하지 않음
    - on=['열이름1','열이름2']: 결합 기준 열이 여러 개일 경우
  - how='outer'(디폴트=inner): 일치하지 않은 결합 기준의 경우 행으로 추가
