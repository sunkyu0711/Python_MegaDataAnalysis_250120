```python
happy_leisure=pd.read_excel('happy\대한민국행복지도_관계및사회참여.xlsx',engine='openpyxl')
happy_leisure.head()
```

```python
import os
path='happy/'
file_list=os.listdir(path)
file_list
```

```python
keywords={'건강':'happy_health','안전':'happy_safe','환경':'happy_environ','경제':'happy_econo','교육':'happy_edu',
          '관계및사회참여':'happy_relation','여가':'happy_leisure'
}

try:
    for i in file_list:
        if i.endswith('.xlsx'):
            for x,y in keywords.items():
                if x in i:
                    keywords[x]=pd.read_excel(path+i,engine='openpyxl')
                    health=keywords[x].groupby('시도')[['평균']].mean()
                    happy_merge=pd.merge(happy_merge,health.rename(columns={'평균':x}),on='시도')
except:
    print(i)
```

```python
happy_merge2=happy_life.groupby(['시도','구군'],as_index=False)[['삶의 만족도']].mean()

keywords={'건강':'happy_health','안전':'happy_safe','환경':'happy_environ','경제':'happy_econo','교육':'happy_edu',
          '관계및사회참여':'happy_relation','여가':'happy_leisure'
}
try:
    for i in file_list:
        if i.endswith('.xlsx'):
            for x in keywords.keys():
                if x in i:
                    keywords[x]=pd.read_excel(path+i,engine='openpyxl')
                    temp=keywords[x].groupby(['시도','구군'],as_index=False)[['평균']].mean()
                    happy_merge2=pd.merge(happy_merge2,temp.rename(columns={'평균':x}))
except:
    print(i)

province_list=list(happy_merge2['시도'])
available=[]
for entry in province_list:
    if entry not in available:
        available.append(entry)
print(f'다음은 조회 가능한 시/도입니다.\n{available}')
print('위 시/도 중 하나를 그대로 입력하지 않으면 오류가 뜨니 주의하시기 바랍니다.')
province=input('조회하고자 하는 시/도를 입력하세요.')
happy_merge2[happy_merge2['시도']==province]
```
