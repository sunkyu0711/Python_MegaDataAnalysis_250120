* 변수명.melt(id_vars=~,var_name=~,value_name=~)
* id_vars: 재구조화의 기준이 되는 열. 한 개든, 여러 개든 리스트를 이용하여 작성할 것
* 단 var_name과 value_name은 하나만을 의미하므로 문자열로 작성

```python
import pandas as pd
```

```python
df = pd.DataFrame({'학생':['파이썬','재미나','판다스'],'수학_2021':[90,85,78],'수학_2022':[95,92,88],'과학_2021':[88,90,85],'과학_2022':[88,90,85]})
df
```

```python
df.melt(id_vars=['학생']) # variable, value 열은 디폴트로 생성
```

```python
df_melt=df.melt(id_vars=['학생'],var_name='과목_연도',value_name='점수')
df_melt
```

```python
df_melt.info()
```

```python
df_melt.head()
```

```python
# 과목과 연도를 구분하고 싶은데?

df_melt['과목_연도'].str.split('_') # 문자열이므로 중간에 str 사용
```

```python
# 리스트로 만들어진 걸 하나의 열로 만들고 싶은데?

df_melt['과목_연도'].str.split('_',expand=True)
```

```python
# 원래 데이터에 합치기

df_melt[['과목','연도']]=df_melt['과목_연도'].str.split('_',expand=True)
df_melt
```

```python
# 기존 과목_연도 열 삭제

df_melt.drop('과목_연도',axis=1,inplace=True)
df_melt
```
