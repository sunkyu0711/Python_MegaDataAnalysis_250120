# 지도 시각화: folium 라이브러리
1. map=folium.Map(): 지도 객체 만들기
2. folium.Marker(속성).add_to(map)
3. 주변 지역을 묶어 표시하기
    - from folium.plugins import MarkerCluster
    - MarkerCluster().add_to(map)

```python
import folium
```

```python
map=folium.Map()
map
```

```python
# 위도 경도 넣기

map=folium.Map(location=[37.4991,127.0285])
map
```

```python
# 확대 축소

map=folium.Map(location=[37.4991,127.0285],zoom_start=15)
map
```

```python
# 지도 크기 조정(픽셀)

map=folium.Map(location=[37.4991,127.0285],zoom_start=20,width=750,height=500)
map
```

```python
# 지점 표시하기

folium.Marker(location=[37.4991,127.0285]).add_to(map)
map
```

```python
# 클릭하면 지점명 표시하기

folium.Marker(location=[37.4991,127.0285],
              popup='메가IT').add_to(map)
map
```

```python
# 지점 표시 꾸미기

folium.Marker(location=[37.4991,127.0285],
              popup='메가IT',
              icon=folium.Icon(color='red',icon='computer',prefix='fa')).add_to(map) # https://fontawesome.com
map
```

## 서울시 반려견 놀이터 지도 시각화

```python
import pandas as pd

df=pd.read_csv('data\서울시_반려견_놀이터.csv')
df
```

```python
df['Latitude'].mean()
```

```python
# 맵 가져오기: 중심은 주소들의 평균으로

import folium

map=folium.Map(location=[df['Latitude'].mean(),df['Longitude'].mean()],zoom_start=10)
map
```

```python
df.index
```

```python
# 각 공원의 위치 표시

df.loc[0,'Latitude']
```

```python
for i in df.index:
    folium.Marker(location=[df.loc[i,'Latitude'],df.loc[i,'Longitude']],
                  popup=df.loc[i,'공원명'],
                  icon=folium.Icon(color='red',icon='dog',prefix='fa')).add_to(map)
map # geocode가 바꾼 객체에 대하여 
```

```python
from folium.plugins import MarkerCluster

map=folium.Map(location=[df['Latitude'].mean(),df['Longitude'].mean()],zoom_start=10)
marker=MarkerCluster().add_to(map)

for i in df.index:
    folium.Marker(location=[df.loc[i,'Latitude'],df.loc[i,'Longitude']],
                  popup=df.loc[i,'공원명'],
                  icon=folium.Icon(color='red',icon='dog',prefix='fa')).add_to(marker)
    
map
```

```python
map.save('map.html')
```
