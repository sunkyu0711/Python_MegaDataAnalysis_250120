```python
# Q4. 관광 비율이 가장 높은/가장 낮은 국가는 어디인가?
# 의문: 아래와 같이 합치면 유학 연수와 기타가 없는 국가도 있는데 비율을 어떻게 따질까? -> pass

kyushu=pd.read_excel('data\구주 입국목적별 자료.xlsx')
tour=pd.read_excel('data\국적별 관광 상용 자료.xlsx')

mg=pd.merge(kyushu,tour,how='outer')
mg
```

```python
# Q5. 관광 목적으로 가장 많이 방문한 주와 국가는 어디인가?

kyushu=pd.read_excel('data\구주 입국목적별 자료.xlsx')
kyushu=kyushu.drop(0,axis=0)
tour=pd.read_excel('data\국적별 관광 상용 자료.xlsx')
tour=tour.drop(0,axis=0)

mg2=pd.merge(kyushu,tour,how='outer')
mg2
```

```python
q5_country=mg2.sort_values(['관광'],ascending=False).head(1)
q5_country[['국적별(2)','관광']]
```
