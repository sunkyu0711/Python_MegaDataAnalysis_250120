## subplot: matplotlib을 이용하여

```python
data={'year':[2010,2011,2012,2013,2014],
      'Sales':[500,800,1200,900,1500],
      'Expenses':[300,400,600,500,800]
}

df2=pd.DataFrame(data)
df2
```

```python
# 한 도화지 안에 4개의 subplot 그릴 거임: plt.subplots(행,열,(각각의 크기))

fig,axes=plt.subplots(2,2,figsize=(10,9))
plt.show()
```

```python
# 프로그래밍적 사고: 그래프는 예쁘게 꾸미기보다는 경향성을 파악하는 데만 주안점을 둠.
# matplotlib 이용
# 첫 번째(0,0)는 연도별 세일에 대한 꺾은선그래프

import pandas as pd
import matplotlib.pyplot as plt

fig,axes=plt.subplots(2,2,figsize=(10,9)) # fig는 figsize를, axes는 subplot의 위치를 의미함
axes[0,0].plot(df2['year'],df2['Sales'])

plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(10,9))
axes[0,0].plot(df2['year'],df2['Sales'])
axes[0,0].set_title('Sales')

plt.show()
```

```python
# 두 번째는 막대그래프

fig,axes=plt.subplots(2,2,figsize=(10,9))
axes[0,0].plot(df2['year'],df2['Sales'])
axes[0,0].set_title('Sales')

axes[0,1].bar(df2['year'],df2['Expenses'])
axes[0,1].set_title('Expenses')

plt.show()
```

```python
# 세 번째는 수치형 데이터 축들끼리의 관계를 볼 때 사용하는 산점도 그래프

fig,axes=plt.subplots(2,2,figsize=(10,9))
axes[0,0].plot(df2['year'],df2['Sales'])
axes[0,0].set_title('Sales')

axes[0,1].bar(df2['year'],df2['Expenses'])
axes[0,1].set_title('Expenses')

axes[1,0].scatter(df2['Sales'],df2['Expenses'])
axes[1,0].set_title('Sales Vs. Expenses') # 비용이 늘수록 판매도 느는 경향이라는 결론을 내릴 수 있음: 비례!

plt.show()
```

```python
# 네 번째는 히스토그램

fig,axes=plt.subplots(2,2,figsize=(10,9))
axes[0,0].plot(df2['year'],df2['Sales'])
axes[0,0].set_title('Sales')

axes[0,1].bar(df2['year'],df2['Expenses'])
axes[0,1].set_title('Expenses')

axes[1,0].scatter(df2['Sales'],df2['Expenses'])
axes[1,0].set_title('Sales Vs. Expenses')

axes[1,1].hist(df2['Sales'])
axes[1,1].set_title('Sales Distribution')

plt.show()
```

```python
# 히스토그램 구간 나누기

fig,axes=plt.subplots(2,2,figsize=(10,9))
axes[0,0].plot(df2['year'],df2['Sales'])
axes[0,0].set_title('Sales') # subplot 옵션 설정할 때에는 set_을 먼저 붙여야!

axes[0,1].bar(df2['year'],df2['Expenses'])
axes[0,1].set_title('Expenses')

axes[1,0].scatter(df2['Sales'],df2['Expenses'])
axes[1,0].set_title('Sales Vs. Expenses')

axes[1,1].hist(df2['Sales'],bins=2) # 디폴트: 10
axes[1,1].set_title('Sales Distribution')

plt.show()
```

```python
# 첫 번째 그래프 디테일 추가

fig,axes=plt.subplots(2,2,figsize=(10,9))

axes[0,0].plot(df2['year'],df2['Sales'])
axes[0,0].set_title('Sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)

axes[0,1].bar(df2['year'],df2['Expenses'])
axes[0,1].set_title('Expenses')

axes[1,0].scatter(df2['Sales'],df2['Expenses'])
axes[1,0].set_title('Sales Vs. Expenses')

axes[1,1].hist(df2['Sales'],bins=2)
axes[1,1].set_title('Sales Distribution')

plt.show()
```

```python
# subplot 사이 간격 조정

fig,axes=plt.subplots(2,2,figsize=(10,9))

axes[0,0].plot(df2['year'],df2['Sales'])
axes[0,0].set_title('Sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)

axes[0,1].bar(df2['year'],df2['Expenses'])
axes[0,1].set_title('Expenses')

axes[1,0].scatter(df2['Sales'],df2['Expenses'])
axes[1,0].set_title('Sales Vs. Expenses')

axes[1,1].hist(df2['Sales'],bins=2)
axes[1,1].set_title('Sales Distribution')

plt.tight_layout()
plt.show()
```

```python
# year가 정수로 안 나옴: 꺾은선(숫자로 인식) vs 막대(문자열로 인식)

fig,axes=plt.subplots(2,2,figsize=(10,9))

axes[0,0].plot(df2['year'],df2['Sales'])
axes[0,0].set_title('Sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)
axes[0,0].set_xticks(df2['year'])

axes[0,1].bar(df2['year'],df2['Expenses'])
axes[0,1].set_title('Expenses')

axes[1,0].scatter(df2['Sales'],df2['Expenses'])
axes[1,0].set_title('Sales Vs. Expenses')

axes[1,1].hist(df2['Sales'],bins=2)
axes[1,1].set_title('Sales Distribution')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data={'year':[2010,2011,2012,2013,2014],
      'Sales':[500,800,1200,900,1500],
      'Expenses':[300,400,600,500,800]
}

df2=pd.DataFrame(data)

fig,axes=plt.subplots(2,2,figsize=(10,9))

plt.show()
```

```python
# 다시, 첫 번째 그래프부터

fig,axes=plt.subplots(2,2,figsize=(10,9))

sns.lineplot(data=df2,x='year',y='Sales',ax=axes[0,0])
axes[0,0].set_title('Sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)

plt.show()
```

```python
# 두 번째 막대그래프

fig,axes=plt.subplots(2,2,figsize=(10,9))

sns.lineplot(data=df2,x='year',y='Sales',ax=axes[0,0])
axes[0,0].set_title('Sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)

sns.barplot(data=df2,x='year',y='Expenses',ax=axes[0,1])

plt.show()
```

```python
# 다채롭게 꾸미기

fig,axes=plt.subplots(2,2,figsize=(10,9))

sns.lineplot(data=df2,x='year',y='Sales',ax=axes[0,0])
axes[0,0].set_title('Sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)

sns.barplot(data=df2,x='year',y='Expenses',ax=axes[0,1],palette='Set1')

plt.show()
```

```python
# 세 번째 산점도

fig,axes=plt.subplots(2,2,figsize=(10,9))

sns.lineplot(data=df2,x='year',y='Sales',ax=axes[0,0])
axes[0,0].set_title('Sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)

sns.barplot(data=df2,x='year',y='Expenses',ax=axes[0,1],palette='Set1')

sns.scatterplot(data=df2,x='Sales',y='Expenses',ax=axes[1,0])

plt.show()
```

```python
# 네 번째 히스토그램

fig,axes=plt.subplots(2,2,figsize=(10,9))

sns.lineplot(data=df2,x='year',y='Sales',ax=axes[0,0])
axes[0,0].set_title('Sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)

sns.barplot(data=df2,x='year',y='Expenses',ax=axes[0,1],palette='Set1')

sns.scatterplot(data=df2,x='Sales',y='Expenses',ax=axes[1,0])

sns.histplot(data=df2,x='Sales',ax=axes[1,1])

plt.show()
```
