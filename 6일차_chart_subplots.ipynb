```python
import pandas as pd
```

```python
data={'이름':['강감찬','홍길동','이순신','최영'],'점수':[50,60,80,65]}

df=pd.DataFrame(data)
df
```

```python
# pandas의 plot 이용한 차트 작성: df.plot()
import matplotlib.pyplot as plt # 한글 포함

df.plot()
```

```python
plt.rc('font', family='Malgun Gothic')
df.plot()
```

```python
df.plot(x='이름')
```

```python
# x축이 범주형 데이터면 막대그래프를 주로 그림

df.plot(kind='bar',x='이름')
```

```python
df.plot(kind='bar',x='이름',rot=0)
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    plt.text(x=i,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점') # 해결!
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점') # 중앙으로 이동! 수동으로 해야 하는 게 단점...
```

```python
# 점수가 전체에서 얼마큼의 비율이 되는지

df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    rate=(df.loc[i]['점수']/df['점수'].sum())*100
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.1,y=df.loc[i]['점수']/2,s=rate)
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%' # 소수점 너무 많은데? 해결!
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.1,y=df.loc[i]['점수']/2,s=rate)
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate) # 정렬 이슈 해결!
```

```python
# 막대 색깔 바꾸기!

df.plot(kind='bar',x='이름',rot=0,ylim=[0,100],color='y')
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%' # 만약 이 라인에서 오류가 난다면
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate) # f-string을 여기에 적용해 보자
```

```python
# 그래프에 평균 라인 그리기!

meanval=df['점수'].mean()
meanval
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100],color='y')
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate)

plt.axhline(y=meanval)
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100],color='y')
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate)

plt.axhline(y=meanval,color='r',linestyle='dashed')
```

```python
# 점수와 평균선이 겹치지 않게 하기

df.plot(kind='bar',x='이름',rot=0,ylim=[0,100],color='y')
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1,y=df.loc[i]['점수']-8,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate)

plt.axhline(y=meanval,color='r',linestyle='dashed')
```

```python
# 평균 점수 보이게 하기

df.plot(kind='bar',x='이름',rot=0,ylim=[0,100],color='y')
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1,y=df.loc[i]['점수']-8,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate)

plt.axhline(y=meanval,color='r',linestyle='dashed')
plt.text(x=0,y=meanval+1,s=meanval)
```

```python
# 좀 더 꾸며 보자

df.plot(kind='bar',x='이름',rot=0,ylim=[0,100],color='y')
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1,y=df.loc[i]['점수']-8,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate)

plt.axhline(y=meanval,color='r',linestyle='dashed')
plt.text(x=0,y=meanval+1,s=f'평균: {meanval:.1f}점')
```

```python
# 위치 이동

df.plot(kind='bar',x='이름',rot=0,ylim=[0,100],color='y')
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1,y=df.loc[i]['점수']-8,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate)

plt.axhline(y=meanval,color='r',linestyle='dashed')
plt.text(x=-0.5,y=meanval+1,s=f'평균: {meanval:.1f}점')
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100],color='y')
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1,y=df.loc[i]['점수']-8,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate)

plt.axhline(y=meanval,color='r',linestyle='dashed')
plt.text(x=-0.5,y=meanval+1,s=f'평균: {meanval:.1f}점')
plt.xlabel('학생 이름')
plt.ylabel('점수')
plt.title('학생별 시험 점수')
```

```python
# 위 그래프를 파일로 저장하기!

df.plot(kind='bar',x='이름',rot=0,ylim=[0,100],color='y')
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1,y=df.loc[i]['점수']-8,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate)

plt.axhline(y=meanval,color='r',linestyle='dashed')
plt.text(x=-0.5,y=meanval+1,s=f'평균: {meanval:.1f}점')

plt.xlabel('학생 이름')
plt.ylabel('점수')
plt.title('학생별 시험 점수')

plt.savefig('graph01.png',dpi=400) # dpi parameter를 바꾸면 size가 달라짐
print('파일이 저장되었습니다.')
plt.show()
```
