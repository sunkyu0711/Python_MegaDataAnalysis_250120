```python
import pandas as pd
```

```python
data={'이름':['강감찬','홍길동','이순신','최영'],'점수':[50,60,80,65]}

df=pd.DataFrame(data)
df
```

```python
# pandas의 plot 이용한 차트 작성: df.plot()
import matplotlib.pyplot as plt # 한글 포함

df.plot()
```

```python
plt.rc('font', family='Malgun Gothic')
df.plot()
```

```python
df.plot(x='이름')
```

```python
# x축이 범주형 데이터면 막대그래프를 주로 그림

df.plot(kind='bar',x='이름')
```

```python
df.plot(kind='bar',x='이름',rot=0)
```

# 6일 차

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
```

```python
# 그래프 위의 데이터 보이게 하려면? pandas에는 그런 거 없어요~
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
plt.text(x=0,y=60,s='값') # 값을 넣을 좌표를 설정하고, 어떤 값을 넣을지
```

```python
df.index
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    plt.text(x=i,y=df.loc[i]['점수'],s=df.loc[i]['점수']) # 막대 높이에 딱 맞춰 값이 쓰임
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    plt.text(x=i,y=df.loc[i]['점수']+1,s=df.loc[i]['점수']) # 해결!
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    plt.text(x=i,y=df.loc[i]['점수']+1,s=df.loc[i]['점수']+'점') # 숫자와 문자열은 +로 연결 불가
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    plt.text(x=i,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점') # 해결!
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점') # 중앙으로 이동! 수동으로 해야 하는 게 단점...
```

```python
# 점수가 전체에서 얼마큼의 비율이 되는지

df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    rate=(df.loc[i]['점수']/df['점수'].sum())*100
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.1,y=df.loc[i]['점수']/2,s=rate)
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%' # 소수점 너무 많은데? 해결!
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.1,y=df.loc[i]['점수']/2,s=rate)
```

```python
df.plot(kind='bar',x='이름',rot=0,ylim=[0,100])
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate) # 정렬 이슈 해결!
```

```python
# 막대 색깔 바꾸기!

df.plot(kind='bar',x='이름',rot=0,ylim=[0,100],color='y')
for i in list(df.index):
    rate=f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1,y=df.loc[i]['점수']+1,s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15,y=df.loc[i]['점수']/2,s=rate)
```
